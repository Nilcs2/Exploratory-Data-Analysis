```python
import pandas as pd
import numpy as np

df = pd.read_csv("titanic.csv")

print(df.isnull().sum())
```

```python
# Sample dataset

data = {"Name": ["Alice", "Bruce", "Bella", np.nan, "David"],
       "Age": [np.nan, 11, 9, 30, 36],
       "Score": [85, 90, np.nan, 91, 88]}
df = pd.DataFrame(data)

print("Original Dataset: \n", df)

df["Age"] = df["Age"].fillna(df["Age"].mean()) # using mean to replace NaN value
df["Score"] = df["Score"].interpolate() # Using interpolate

print("\nDataset: \n", df)

# Renaming columns and rows

df = df.rename(columns={"Name": "Student_Name", "Score": "Exam_score"})

print("\nRenamed dataset:\n", df)
```

```python
# Merge 2 datasets and merge them

df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Bruce", "Bella", "Buddy"],
    "Age": [11, 10, 4]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Score": [85, 90, 88]
})

print("Dataset: \n", df1)
print("\nDataset: \n", df2)

merged_dataset = pd.merge(df1, df2, how="inner", on="ID")
print("\nMerged Dataset: \n", merged_dataset)
merged_dataset["Score_Percentage"] = (merged_dataset["Score"] / 100) * 100
print("\nTransformed Dataset: \n", merged_dataset)
```
