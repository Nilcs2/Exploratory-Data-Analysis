```python
# Group data by categorical column

import pandas as pd

data = {
    "Grade": ["A", "B", "C", "A", "C", "B"],
    "Score": [90, 89, 75, 91, 79, 85],
    "Age": [35, 23, 23, 27, 22, 28],
}

df = pd.DataFrame(data)

print("Original Dataset:\n", df)
```

```python
grouped = df.groupby("Grade").mean()
print(grouped)
```

```python
# Calculate Summary statistics for grouped data
stats = df.groupby("Grade").agg({"Score": ["mean", "max", "min"], "Age": ["mean", "max", "min"]})
print(stats)
```
